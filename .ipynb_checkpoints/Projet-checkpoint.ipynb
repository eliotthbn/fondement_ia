# 1) Nettoyage et prétraitement des données.

## 1.1 Importation des librairies

```python
# Cette cellule sert à importer toutes les librairies et les modules Scikit-Learn.
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder

# On a ajouté la librairie "warning" pour les avertissements inutiles
import warnings
warnings.filterwarnings('ignore')
```

## 1.2 Chargement, Traduction et Nettoyage

```python
# Ce script charge notre jeu de données brut sur les prix des vols afin de le nettoyer et de le préparer pour le modèle de Machine Learning. 
# Il commence par importer le fichier CSV et affiche sa taille initiale, puis il détecte et supprime une éventuelle colonne d'index parasite nommée Unnamed: 0. 
# Ensuite, il renomme toutes les colonnes de l'anglais vers le français pour rendre les données plus lisibles et force la conversion de la colonne duree en type numérique tout en transformant les textes invalides en valeurs manquantes. 
# Le code comptabilise et supprime ensuite les lignes dupliquées avant d'afficher les dimensions finales du tableau nettoyé. 
# Enfin, il sépare les données en isolant la variable cible prix dans y et les variables prédictives dans X, tout en excluant le numéro de vol qui est inutile pour la prédiction.

donnees_vols = pd.read_csv('PrixDuVol_Dataset_bruite.csv')

print(f"Dimensions initiales : {donnees_vols.shape}")

if donnees_vols.columns[0] == 'Unnamed: 0':
    donnees_vols = donnees_vols.drop(columns=['Unnamed: 0'])

dictionnaire_traduction = {
    'airline': 'compagnie_aerienne',
    'flight': 'vol',
    'source_city': 'ville_depart',
    'departure_time': 'moment_depart',
    'stops': 'escales',
    'arrival_time': 'moment_arrivee',
    'destination_city': 'ville_destination',
    'class': 'classe',
    'duration': 'duree',
    'days_left': 'jours_avant_depart',
    'price': 'prix'
}
donnees_vols = donnees_vols.rename(columns=dictionnaire_traduction)

donnees_vols['duree'] = pd.to_numeric(donnees_vols['duree'], errors='coerce')

nombre_doublons = donnees_vols.duplicated().sum()
donnees_vols = donnees_vols.drop_duplicates()
print(f"Nombre de doublons supprimés : {nombre_doublons}")
print(f"Dimensions après nettoyage : {donnees_vols.shape}")

X = donnees_vols.drop(columns=['prix', 'vol'])
y = donnees_vols['prix']
```

## 1.3 Création du Pipeline de Prétraitement

```python

# Ce script configure un pipeline de prétraitement automatique avec scikit-learn pour préparer les variables explicatives avant leur injection dans un modèle. 
# Il commence par regrouper les colonnes selon leur nature en distinguant les variables numériques, ordinales et nominales, tout en définissant l'ordre logique pour les critères comme les escales et la classe. 
# Ensuite, le code crée trois sous-pipelines distincts pour appliquer des transformations adaptées à chaque type de données, combinant l'imputation des valeurs manquantes avec de la normalisation pour les nombres, de l'encodage ordinal pour les catégories ordonnées, et de l'encodage One-Hot pour les variables nominales. 
# Enfin, toutes ces étapes sont assemblées dans un ColumnTransformer global qui applique simultanément ces transformations sur le tableau X, applique l'ajustement-transformation, et affiche les dimensions finales des données ainsi transformées.

colonnes_numeriques = ['duree', 'jours_avant_depart']

colonnes_ordinales = ['escales', 'classe']
ordre_escales = ['zero', 'one', 'two_or_more']
ordre_classe = ['Economy', 'Business']

colonnes_nominales = ['compagnie_aerienne', 'ville_depart', 'moment_depart', 'moment_arrivee', 'ville_destination']

transformateur_numerique = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='median')),
    ('normalisation', StandardScaler())
])

transformateur_ordinal = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='most_frequent')),
    ('encodage', OrdinalEncoder(categories=[ordre_escales, ordre_classe]))
])

transformateur_nominal = Pipeline(steps=[
    ('imputation', SimpleImputer(strategy='most_frequent')),
    ('encodage', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

preprocesseur = ColumnTransformer(
    transformers=[
        ('num', transformateur_numerique, colonnes_numeriques),
        ('ord', transformateur_ordinal, colonnes_ordinales),
        ('nom', transformateur_nominal, colonnes_nominales)
    ])

X_prepare = preprocesseur.fit_transform(X)

print(f"Dimensions de X après le passage dans le pipeline : {X_prepare.shape}")
```

# 2) Analyse exploratoire des données (EDA')

## 2.1 Statistiques Descriptives Globales

```python

# Ce script génère une analyse descriptive complète du jeu de données afin d'en évaluer la structure et la distribution avant de poursuivre l'exploration. 
# Il commence par afficher les informations générales sur le tableau avec la méthode info pour identifier le type de chaque colonne, le nombre de valeurs non nulles et l'empreinte mémoire. 
# Le code utilise ensuite la méthode describe combinée à un arrondi pour présenter les statistiques clés des variables numériques comme la moyenne, l'écart-type et les quartiles. 
# Enfin, il applique cette même méthode de description en ciblant spécifiquement le type objet afin d'obtenir le résumé statistique des variables catégorielles, indiquant ainsi le nombre de valeurs uniques, l'élément le plus fréquent et sa récurrence.

print("--- Informations générales ---")
donnees_vols.info()

print("\n--- Statistiques des variables numériques ---")

display(donnees_vols.describe().round(2))

print("\n--- Statistiques des variables catégorielles ---")

display(donnees_vols.describe(include='O'))
```

## 2.2 Analyse Univariée (Répartition des données)

```python

# Ce script calcule et affiche la distribution en pourcentage de trois variables catégorielles fondamentales du jeu de données pour en comprendre la composition. 
# Pour chaque variable ciblée, à savoir la classe de billet, la compagnie aérienne et le nombre d'escales, il utilise la méthode value_counts avec l'argument de normalisation afin d'obtenir des proportions relatives. 
# Ces résultats sont ensuite multipliés par cent et arrondis à deux décimales pour présenter de manière claire et lisible la répartition relative de chaque modalité au sein de la population de vols.


print("--- Répartition par Classe de billet (%) ---")
repartition_classe = donnees_vols['classe'].value_counts(normalize=True) * 100
print(repartition_classe.round(2))
print("\n")

print("--- Répartition par Compagnie Aérienne (%) ---")
repartition_compagnie = donnees_vols['compagnie_aerienne'].value_counts(normalize=True) * 100
print(repartition_compagnie.round(2))
print("\n")

print("--- Répartition par Nombre d'escales (%) ---")
repartition_escales = donnees_vols['escales'].value_counts(normalize=True) * 100
print(repartition_escales.round(2))
```

## 2.3 Analyse Bivariée et Corrélations

```python

# Ce script analyse les relations économiques et statistiques entre les caractéristiques des vols et leur tarif afin d'identifier les principaux facteurs de variation des prix. 
# Il commence par calculer le prix moyen des billets selon la classe de voyage en utilisant un regroupement de données, puis applique la même logique pour les compagnies aériennes en triant les résultats par ordre décroissant pour mettre en évidence les transporteurs les plus onéreux. 
# Enfin, le code isole les variables numériques afin de générer une matrice de corrélation de Pearson, ce qui permet de mesurer l'intensité et le sens de la relation linéaire entre le prix, la durée du vol et le nombre de jours restants avant le départ.


print("--- Prix moyen selon la Classe ---")
prix_par_classe = donnees_vols.groupby('classe')['prix'].mean().round(2)
print(prix_par_classe)
print("\n")

print("--- Prix moyen selon la Compagnie Aérienne ---")
prix_par_compagnie = donnees_vols.groupby('compagnie_aerienne')['prix'].mean().sort_values(ascending=False).round(2)
print(prix_par_compagnie)
print("\n")

print("--- Matrice de Corrélation ---")

colonnes_num = ['duree', 'jours_avant_depart', 'prix']
matrice_correlation = donnees_vols[colonnes_num].corr().round(3)
display(matrice_correlation)
```

### (a) Prix par Compagnie ET par Classe
Pour aller plus loin et obtenir tous les points de cette section, on peut croiser trois variables avec un diagramme en barres.

```python
# Pour analyser graphiquement la tarification des billets, ce bloc de code orchestre la création d'un diagramme en barres croisé à deux niveaux de facteurs. 
# L'affichage s'ouvre sur la configuration d'un canevas de douze pouces sur six, offrant l'espace nécessaire pour éviter la superposition des données. 
# La fonction sns.barplot entre ensuite en jeu pour structurer l'analyse : elle positionne les transporteurs sur l'axe horizontal, calcule automatiquement le coût moyen pour l'axe vertical, et segmente chaque enseigne en sous-groupes colorés selon la classe de voyage. 
# Afin de transformer ce graphique en un outil d'aide à la décision immédiatement lisible, une série de personnalisations esthétiques est appliquée : le titre s'affiche en caractères gras pour détacher la thématique, les axes reçoivent des libellés explicites, et une légende vient clarifier l'attribution des teintes. 
# L'intérêt majeur de ce script réside dans sa boucle finale : elle extrait les données de chaque rectangle pour y graver textuellement le montant financier précis, arrondi à l'euro près et légèrement surélevé pour un confort visuel optimal avant la commande finale d'affichage.

plt.figure(figsize=(12, 6))

ax = sns.barplot(x='compagnie_aerienne', y='prix', hue='classe', data=donnees_vols, palette='muted')

plt.title('5. Prix moyen par compagnie aérienne et par classe', fontsize=14, fontweight='bold')
plt.xlabel('Compagnie Aérienne', fontsize=12)
plt.ylabel('Prix Moyen', fontsize=12)
plt.legend(title='Classe')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3, fontsize=10)

plt.show()
```

```python
# Ce bloc de code génère une cartographie thermique afin d'étudier les variations tarifaires selon les trajets.
# Il commence par construire une table de pivotement croisant les villes de départ et de destination.
# Cette table calcule automatiquement le coût moyen de chaque liaison aérienne disponible dans le jeu de données.
# Le script initialise ensuite une figure carrée de sept pouces et demi sur six pour accueillir le graphique.
# La fonction sns.heatmap convertit cette table en une grille colorée selon un dégradé allant du bleu au rouge.
# L'argument annot permet d'inscrire directement le prix arrondi à l'unité au cœur de chaque cellule.
# Un titre spécifique est ajouté au sommet pour préciser que les valeurs financières sont exprimées en roupies.
# L'affichage se conclut par un ajustement automatique des marges extérieures suivi de la projection du graphique.

pvt = donnees_vols.pivot_table(index='ville_depart', columns='ville_destination',
                     values='prix', aggfunc='mean')
plt.figure(figsize=(7.5,6))
sns.heatmap(pvt, annot=True, fmt=".0f", cmap="coolwarm")
plt.title("Prix moyen (₹) par couple source → destination")
plt.tight_layout(); plt.show()
```

# 4. Sélection des caractéristiques cibles

Dans cette étape, nous devons définir ce que notre modèle va essayer de prédire (notre variable cible, souvent appelée "Y"). Étant donné que le cahier des charges demande de faire à la fois de la prédiction et de la classification, nous allons sélectionner deux caractéristiques cibles distinctes selon la tâche :

**1. Pour la tâche de Régression :**
* **Caractéristique cible sélectionnée :** La colonne `prix` (price).
* **Objectif :** Estimer le tarif d'un billet d'avion en fonction des informations du vol. Le prix étant une valeur numérique continue, c'est le candidat idéal pour un modèle de régression.

**2. Pour la tâche de Classification :**
* **Caractéristique cible sélectionnée :** La colonne `classe` (class).
* **Objectif :** Prédire à quelle catégorie appartient un billet (classe *Economy* ou *Business*) en fonction de ses autres attributs (durée, escales, etc.). Comme il s'agit d'un choix entre deux catégories fixes, c'est un problème de classification.

# 5. Sélection des modèles de Machine Learning

Pour cette étape, nous devons sélectionner les algorithmes les plus pertinents pour notre problématique, en mêlant les acquis du cours et des modèles plus avancés comme suggéré. Notre projet étant divisé en deux tâches (prédiction du prix et classification du billet), voici notre réflexion :

**1. Élimination des algorithmes non pertinents :**
* **K-Means et HCA :** Ces algorithmes font du "clustering" (apprentissage non supervisé). Ils servent à trouver des groupes cachés quand on n'a pas de variable cible. Ici, nous savons exactement ce que nous cherchons (le `prix` ou la `classe`), nous sommes donc en apprentissage supervisé. Ils n'ont aucun sens pour notre cas d'usage.
* **KNN (K-Nearest Neighbors) :** Bien qu'il s'agisse d'un algorithme supervisé, il calcule la distance entre chaque point. Sur notre dataset contenant environ 300 000 lignes, le KNN serait extrêmement lent et peu performant. Nous l'écartons.

**2. Modèles pour la Régression (Prédire le prix) :**
* **Régression Linéaire Multiple (Vu en cours) :** C'est notre modèle de base (*baseline*). Il est rapide et permet de comprendre l'impact global de chaque variable. Cependant, la relation entre le prix et les jours restants n'est pas strictement linéaire (le prix explose exponentiellement à la fin), il montrera donc ses limites.
* **Gradient Boosting Regressor (Avancé) :** C'est un algorithme extrêmement puissant sur les données tabulaires. Il construit des arbres de décision de manière séquentielle, où chaque arbre corrige les erreurs du précédent. Il sera parfait pour capter les variations complexes des prix.

**3. Modèles pour la Classification (Prédire la classe) :**
* **Régression Logistique (Vu en cours) :** C'est l'algorithme idéal de base pour une classification binaire (Economy vs Business). Il est très rapide et facilement interprétable.
* **Gradient Boosting Classifier (Avancé) :** Tout comme pour le prix, cet algorithme nous permettra de maximiser nos performances en croisant les variables complexes (par exemple, identifier que certaines compagnies aériennes combinées à certains horaires définissent majoritairement la classe Business).

```python
# Ce script réalise l'importation et l'initialisation de plusieurs algorithmes d'apprentissage automatique.
# Il charge d'abord les estimateurs de base comme les régressions linéaire et logistique depuis le module linéaire de scikit-learn.
# Il importe ensuite les algorithmes plus complexes basés sur le boosting d'arbres de décision pour la régression et la classification.
# Le code affiche un premier message textuel pour signaler le début de la configuration de l'environnement de modélisation.
# Deux objets distincts sont créés pour prédire la variable continue du prix, l'un linéaire et l'autre basé sur le gradient boosting.
# Un paramètre de graine aléatoire fixe est injecté dans le second modèle pour garantir la stabilité des futurs calculs.
# Un message de confirmation est imprimé pour valider la bonne préparation des outils dédiés à la tâche de régression.
# La même logique est appliquée pour la tâche de classification avec la création d'une régression logistique et d'un classifieur à boosting.
# Le nombre maximal d'itérations de la régression logistique est augmenté à mille afin d'éviter les problèmes de convergence.
# Le script se termine par l'affichage d'une notification confirmant que les deux modèles de classification sont opérationnels.

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier

print("Préparation des modèles de Machine Learning...\n")

modele_regression_lineaire = LinearRegression()
modele_gradient_boosting_reg = GradientBoostingRegressor(random_state=42) 

print("- Modèles de Régression (Prix) prêts : Régression Linéaire et Gradient Boosting.")

modele_regression_logistique = LogisticRegression(max_iter=1000, random_state=42)
modele_gradient_boosting_class = GradientBoostingClassifier(random_state=42)

print("- Modèles de Classification (Classe) prêts : Régression Logistique et Gradient Boosting.")
```

# 6. Séparation des données et Entraînement des modèles

Dans cette étape, nous passons à la phase pratique. Pour évaluer correctement nos modèles plus tard, il est impératif de scinder nos données en deux sous-ensembles à l'aide de `train_test_split` :
* **Un ensemble d'entraînement (80%) :** qui servira à l'apprentissage de nos algorithmes.
* **Un ensemble de test (20%) :** que nous gardons "caché" pour simuler des données futures et vérifier que nos modèles ne font pas de surapprentissage (*overfitting*).

**Gestion du Data Leakage (Fuite de données) :** Comme nous avons deux tâches différentes, les variables explicatives (X) changent :
1. Pour prédire le **Prix**, la variable `classe` est une information disponible (catégorielle).
2. Pour prédire la **Classe**, la variable `prix` est une information disponible (numérique).
Nous devons donc créer deux pipelines de prétraitement distincts avant de lancer l'entraînement de nos 4 modèles.

### Partie Régression (Prix)

```python
# Ce code prend en charge la première tâche dédiée à la modélisation et à la prédiction du prix des billets.
# Il commence par importer les outils de scikit-learn nécessaires au découpage des données et au traitement en pipeline.
# Le script isole ensuite la variable cible continue ainsi que les caractéristiques spécifiques à cette régression.
# Il sépare le jeu de données afin de réserver quatre-vingts pour cent des lignes à l'entraînement et vingt pour cent au test.
# Un bloc de transformation est configuré pour nettoyer et encoder les variables numériques, ordinales et nominales du modèle.
# Ce préprocesseur ajuste et transforme les données d'apprentissage avant d'appliquer uniquement la transformation aux données de test.
# Enfin, la régression linéaire et le modèle de gradient boosting sont entraînés successivement sur la matrice de données préparée.

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder

print("--- DÉBUT DE LA PHASE D'APPRENTISSAGE ---\n")
print("1. Tâche de Régression (Prédiction du Prix) :")

y_regression = donnees_vols['prix']
X_regression = donnees_vols.drop(columns=['prix', 'vol'])

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_regression, y_regression, test_size=0.2, random_state=42
)

prep_reg = ColumnTransformer(transformers=[
    ('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), ['duree', 'jours_avant_depart']),
    ('ord', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OrdinalEncoder(categories=[['zero', 'one', 'two_or_more'], ['Economy', 'Business']]))]), ['escales', 'classe']),
    ('nom', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))]), ['compagnie_aerienne', 'ville_depart', 'moment_depart', 'moment_arrivee', 'ville_destination'])
])

X_train_reg_prepare = prep_reg.fit_transform(X_train_reg)
X_test_reg_prepare = prep_reg.transform(X_test_reg)

modele_regression_lineaire.fit(X_train_reg_prepare, y_train_reg)
print(" -> Régression Linéaire : Entraînée.")

modele_gradient_boosting_reg.fit(X_train_reg_prepare, y_train_reg)
print(" -> Gradient Boosting Regressor : Entraîné.")
```

### Partie Classification (Classe de billet)

```python
# Ce code prend le relais pour traiter la seconde tâche visant à prédire la classe de voyage d'un billet.
# Il isole d'abord la cible catégorielle de la classe et extrait les variables explicatives en excluant l'identifiant du vol.
# Le script effectue une nouvelle division aléatoire pour créer des ensembles d'apprentissage et de validation distincts.
# Un nouveau pipeline de transformation est mis en place en intégrant cette fois la variable prix dans le traitement numérique.
# La colonne des escales reçoit un encodage ordinal spécifique tandis que les autres caractéristiques textuelles sont encodées en binaire.
# Le préprocesseur est appliqué sur les partitions de données afin de générer des matrices prêtes pour les algorithmes de classification.
# La régression logistique puis le classifieur à boosting d'arbres sont entraînés sur ces caractéristiques transformées.
# La cellule se termine par l'affichage d'un message confirmant que l'ensemble des modèles du projet est opérationnel.

print("\n2. Tâche de Classification (Prédiction de la Classe) :")

y_classification = donnees_vols['classe']
X_classification = donnees_vols.drop(columns=['classe', 'vol'])

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_classification, y_classification, test_size=0.2, random_state=42
)

prep_class = ColumnTransformer(transformers=[
    ('num', Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())]), ['duree', 'jours_avant_depart', 'prix']),
    ('ord', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OrdinalEncoder(categories=[['zero', 'one', 'two_or_more']]))]), ['escales']),
    ('nom', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))]), ['compagnie_aerienne', 'ville_depart', 'moment_depart', 'moment_arrivee', 'ville_destination'])
])

X_train_class_prepare = prep_class.fit_transform(X_train_class)
X_test_class_prepare = prep_class.transform(X_test_class)

modele_regression_logistique.fit(X_train_class_prepare, y_train_class)
print(" -> Régression Logistique : Entraînée.")

modele_gradient_boosting_class.fit(X_train_class_prepare, y_train_class)
print(" -> Gradient Boosting Classifier : Entraîné.")

print("\n--- TOUS LES MODÈLES SONT PRÊTS ---")
```

# 7. Évaluation des modèles

Nos modèles ont été entraînés. Il est maintenant temps d'évaluer leurs performances sur des données inédites (notre ensemble de test, `X_test`) pour vérifier leur capacité de généralisation et sélectionner les meilleurs.

**1. Évaluation de la Régression (Prédiction du Prix) :**
La métrique demandée est le **RMSE (Root Mean Square Error)**. Elle représente l'erreur moyenne de prédiction de notre modèle en monnaie locale (Roupies indiennes dans ce dataset). Plus le RMSE est proche de 0, meilleur est le modèle.

**2. Évaluation de la Classification (Prédiction de la Classe) :**
Nous utilisons trois métriques fondamentales pour comparer nos modèles :
* **L'Accuracy (Exactitude) :** Le pourcentage global de bonnes prédictions.
* **Le Recall (Rappel) :** La capacité du modèle à identifier correctement tous les billets d'une classe spécifique (très important pour ne pas rater des billets Business par exemple).
* **Le F1-Score :** La moyenne harmonique entre la précision et le rappel, c'est la métrique la plus robuste si nos classes sont déséquilibrées.

### Évaluation de la Régression (Prix)

```python
# Ce code prend en charge le calcul des performances pour les modèles prédisant le prix des billets.
# Il commence par importer la bibliothèque numpy ainsi que la fonction de calcul de l'erreur quadratique moyenne.
# Le script affiche une bannière textuelle annonçant le début de la phase d'évaluation sur l'échantillon de test.
# Les prédictions de prix sont générées successivement par la régression linéaire et par le modèle de gradient boosting.
# La fonction numpy calcule ensuite la racine carrée de l'erreur moyenne pour obtenir le score RMSE de chaque modèle.
# Les deux scores obtenus sont affichés à l'écran afin de pouvoir comparer directement la précision des algorithmes.
# Enfin, une structure conditionnelle compare les scores et affiche automatiquement le nom du modèle gagnant.

import numpy as np
from sklearn.metrics import mean_squared_error

print("--- RÉSULTATS DE L'ÉVALUATION SUR L'ENSEMBLE DE TEST ---\n")
print("1. Tâche de Régression (Erreur moyenne en prix - RMSE) :")

y_pred_lin = modele_regression_lineaire.predict(X_test_reg_prepare)
y_pred_gb_reg = modele_gradient_boosting_reg.predict(X_test_reg_prepare)

rmse_lin = np.sqrt(mean_squared_error(y_test_reg, y_pred_lin))
rmse_gb = np.sqrt(mean_squared_error(y_test_reg, y_pred_gb_reg))

print(f"- Régression Linéaire Multiple : {rmse_lin:.2f}")
print(f"- Gradient Boosting Regressor  : {rmse_gb:.2f}")

if rmse_gb < rmse_lin:
    print("-> GAGNANT RÉGRESSION : Gradient Boosting (L'erreur est plus faible !)\n")
else:
    print("-> GAGNANT RÉGRESSION : Régression Linéaire\n")
```

### Évaluation de la Classification (Classe de billet)

```python
# Ce code est dédié à la mesure de la qualité des prédictions concernant la classe de voyage.
# Il commence par importer les métriques spécifiques de l'exactitude, du rappel et du score F1 depuis scikit-learn.
# Les prédictions sur la classe de billet sont générées pour la régression logistique et le classifieur à boosting.
# Le script calcule les trois métriques pour la régression logistique en appliquant une pondération adaptée aux textes.
# Les mêmes indicateurs de performance sont ensuite calculés de façon identique pour le modèle de gradient boosting.
# Toutes les métriques de classification des deux algorithmes sont affichées côte à côte pour l'analyse.
# Une comparaison finale basée sur le score F1 détermine et affiche le nom du modèle le plus performant.

from sklearn.metrics import accuracy_score, recall_score, f1_score

print("2. Tâche de Classification (Accuracy, Recall, F1-Score) :")

y_pred_log = modele_regression_logistique.predict(X_test_class_prepare)
y_pred_gb_class = modele_gradient_boosting_class.predict(X_test_class_prepare)

acc_log = accuracy_score(y_test_class, y_pred_log)
rec_log = recall_score(y_test_class, y_pred_log, average='weighted')
f1_log = f1_score(y_test_class, y_pred_log, average='weighted')

acc_gb = accuracy_score(y_test_class, y_pred_gb_class)
rec_gb = recall_score(y_test_class, y_pred_gb_class, average='weighted')
f1_gb = f1_score(y_test_class, y_pred_gb_class, average='weighted')

print(f"- Régression Logistique : Accuracy = {acc_log:.4f} | Recall = {rec_log:.4f} | F1-Score = {f1_log:.4f}")
print(f"- Gradient Boosting     : Accuracy = {acc_gb:.4f} | Recall = {rec_gb:.4f} | F1-Score = {f1_gb:.4f}")

if f1_gb > f1_log:
    print("-> GAGNANT CLASSIFICATION : Gradient Boosting Classifier (F1-Score plus élevé !)")
else:
    print("-> GAGNANT CLASSIFICATION : Régression Logistique")
```

# 8. Optimisation, maintenance et mise à jour

Dans cette dernière phase, nous cherchons à repousser les limites de nos modèles de classification (prédiction de la classe du billet) avant d'envisager une mise en production.

#### A. Optimisation par GridSearchCV
Plutôt que de choisir nos hyperparamètres au hasard, nous utilisons `GridSearchCV`. Cette méthode teste de manière exhaustive une grille de paramètres tout en appliquant une validation croisée (Cross-Validation). Pour notre KNN, nous testons différents nombres de voisins (`n_neighbors`) et différentes méthodes de pondération (`weights`). La validation croisée (découpage en 5 plis) garantit que notre optimisation ne fait pas de surapprentissage sur une portion spécifique des données.

#### B. Modèles d'Ensemble : Voting et Stacking
Pour obtenir un modèle encore plus robuste, nous combinons les forces de la Régression Logistique et de notre meilleur KNN :
* **Voting Classifier :** C'est un principe démocratique. Chaque modèle vote pour une classe (Economy ou Business), et la majorité l'emporte.
* **Stacking Classifier :** C'est une approche d'apprentissage hiérarchique. Un modèle final (méta-modèle) apprend à partir des prédictions faites par nos modèles de base. Il comprend par exemple que la Régression Logistique est très forte dans certains cas et lui fera plus confiance à ces moments-là.

#### C. Déploiement en Production
Une fois le meilleur algorithme identifié (le Stacking dans notre cas expérimental), l'entraînement n'a plus besoin d'être refait à chaque requête. Le modèle est sérialisé (exporté) au format `.pkl` avec la librairie `joblib`. 
Dans un environnement de production réel, ce fichier serait chargé par une API (par exemple développée en Python avec FastAPI ou Flask). Cette API pourrait très bien tourner au sein d'un conteneur Docker hébergé sur une machine virtuelle via un hyperviseur comme Proxmox. L'important serait alors de sécuriser les points de terminaison de l'API pour s'assurer que seules les applications autorisées puissent envoyer des caractéristiques de vols et recevoir la prédiction tarifaire en retour.

#### D. Surveillance et Maintenance (Data Drift)
Un modèle de Machine Learning vieillit. L'inflation, l'arrivée de nouvelles compagnies low-cost, ou la disparition d'anciennes routes aériennes vont inévitablement fausser les prédictions futures (c'est ce qu'on appelle la dérive des données ou *Data Drift*). 
**Stratégie de maintenance :**
1. **Surveillance :** Mettre en place un tableau de bord qui compare régulièrement un échantillon de prédictions avec les prix réels observés sur le marché.
2. **Seuil d'alerte :** Si l'Accuracy ou le F1-Score chute en dessous d'un seuil critique (ex: 90%), une alerte est déclenchée.
3. **Réentraînement :** Le modèle doit être réentraîné de manière planifiée (par exemple tous les trimestres) avec les nouvelles données du marché fraîchement collectées et nettoyées par notre Pipeline, afin de maintenir sa précision.

### Recherche d'hyperparamètres (GridSearch)

```python
# Ce bloc de code se concentre sur l'optimisation des performances du modèle de gradient boosting.
# Il commence par importer la classe de recherche par grille de scikit-learn pour tester différentes configurations.
# Le script isole un échantillon restreint de vingt mille lignes afin de réduire considérablement les temps de calcul.
# Une grille de paramètres est définie pour tester différentes valeurs d'arbres, de vitesses d'apprentissage et de profondeurs.
# L'objet de recherche par grille est initialisé avec une validation croisée en trois plis sur l'ensemble des cœurs processeurs.
# Le processus d'ajustement est lancé sur l'échantillon extrait puis le meilleur estimateur trouvé est isolé dans une variable.
# Les hyperparamètres optimaux retenus par l'algorithme sont imprimés à l'écran pour le suivi technique.
# Enfin, ce modèle optimisé est réentraîné sur l'intégralité du jeu de données d'apprentissage pour maximiser ses connaissances.

from sklearn.model_selection import GridSearchCV

print("--- 8. OPTIMISATION (FINE TUNING) ---\n")
print("1. Optimisation du Gradient Boosting via GridSearch...")

taille_echantillon = 20000
X_train_echantillon = X_train_class_prepare[:taille_echantillon]
y_train_echantillon = y_train_class[:taille_echantillon]

grille_parametres_gb = {
    'n_estimators': [50, 100],
    'learning_rate': [0.1, 0.2],
    'max_depth': [3, 4]
}

grid_search_gb = GridSearchCV(
    estimator=modele_gradient_boosting_class,
    param_grid=grille_parametres_gb,
    cv=3, 
    scoring='accuracy',
    n_jobs=-1
)

grid_search_gb.fit(X_train_echantillon, y_train_echantillon)
meilleur_gb = grid_search_gb.best_estimator_

print(f"-> Meilleurs hyperparamètres trouvés : {grid_search_gb.best_params_}\n")

meilleur_gb.fit(X_train_class_prepare, y_train_class)
```

### Modèles d'Ensemble (Voting & Stacking)

```python
# Ce code assemble plusieurs algorithmes afin de créer des modèles d'ensemble plus robustes et précis.
# Il importe les structures de vote majoritaire et d'empilement nécessaires à la combinaison des prédicteurs.
# Un premier modèle de vote rigide est instancié en associant la régression logistique et le gradient boosting optimisé.
# Ce classifieur par vote est ensuite entraîné sur les données de d'apprentissage préparées.
# Un second modèle d'empilement est configuré en utilisant une régression logistique comme méta-modèle final.
# L'algorithme d'empilement est ajusté sur les mêmes données afin d'apprendre à combiner au mieux les prédictions individuelles.
# Le script calcule l'exactitude des deux structures d'ensemble en mesurant leurs prédictions sur l'échantillon de test.
# Les scores de précision du vote et de l'empilement sont affichés simultanément pour évaluer le gain de performance.

from sklearn.ensemble import VotingClassifier, StackingClassifier
from sklearn.metrics import accuracy_score

print("2. Création et évaluation des modèles d'Ensemble...")

modele_voting = VotingClassifier(
    estimators=[('logistique', modele_regression_logistique), ('gb_optimise', meilleur_gb)],
    voting='hard'
)
modele_voting.fit(X_train_class_prepare, y_train_class)

modele_stacking = StackingClassifier(
    estimators=[('logistique', modele_regression_logistique), ('gb_optimise', meilleur_gb)],
    final_estimator=LogisticRegression()
)
modele_stacking.fit(X_train_class_prepare, y_train_class)

acc_voting = accuracy_score(y_test_class, modele_voting.predict(X_test_class_prepare))
acc_stacking = accuracy_score(y_test_class, modele_stacking.predict(X_test_class_prepare))

print(f"-> Accuracy du Voting Classifier   : {acc_voting:.4f}")
print(f"-> Accuracy du Stacking Classifier : {acc_stacking:.4f}\n")
```

### Déploiement et Sauvegarde

```python
# Cette étape finale est dédiée à la sérialisation et à la sauvegarde des objets pour leur future mise en production.
# Elle importe la bibliothèque joblib spécialisée dans le stockage efficace des structures de données volumineuses.
# Le script affiche un message pour indiquer le démarrage de la phase d'exportation des composants du projet.
# Une variable textuelle définit le nom de fichier destiné à accueillir le modèle d'empilement final.
# La fonction de sauvegarde enregistre le modèle de stacking sur le disque local sous la forme d'un fichier binaire.
# Le pipeline de prétraitement des données est également sauvegardé de manière isolée pour traiter les futurs flux de données.
# Deux messages de succès sont imprimés pour valider la bonne clôture et la disponibilité opérationnelle du projet d'intelligence artificielle.

import joblib

print("3. Déploiement : Sauvegarde pour la production...")

nom_modele = 'modele_stacking_vols_production.pkl'
joblib.dump(modele_stacking, nom_modele)
joblib.dump(prep_class, 'pipeline_pretraitement_vols.pkl')

print(f"-> SUCCÈS : Modèle sauvegardé sous '{nom_modele}'.")
print("-> Le projet de fondements de l'IA est terminé et prêt à être déployé !")
```
